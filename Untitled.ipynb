# TrainTest - Netoyage - Encodage

```python
from sklearn.preprocessing import OrdinalEncoder, OneHotEncoder
from sklearn.model_selection import train_test_split
```

```python
train_set, test_set = train_test_split(df, test_size=0.2)
train_set
```

```python
def imputation(df):
#     df = df.dropna()
    for col in df.columns:
        df[col] = df[col].fillna(df[col].value_counts().idxmax())
    
    return df
```

```python
def encodage(df):
    df = imputation(df)
    for col in df.select_dtypes("object").columns:
        encoder = OrdinalEncoder()
        df[col] = encoder.fit_transform(df[[col]])
        
    return df
```

```python
def preprocessing(df):
    df = encodage(df)
    
    X = df.drop("SalePrice", axis=1)
    y = df["SalePrice"]
    
    return X, y
```

```python
def preprocessingForTest(df):
    df = encodage(df)
    
    return df
```

```python
X_train, y_train = preprocessing(train_set)
```

```python
X_test, y_test = preprocessing(test_set)
```

# Modélisation

```python
from sklearn.linear_model import SGDRegressor
from sklearn.pipeline import make_pipeline
from sklearn.preprocessing import StandardScaler
```

```python
SGD = make_pipeline(StandardScaler(), SGDRegressor(max_iter=1000, tol=1e-3))
```

# Procédure d'évaluation

```python
from sklearn.metrics import f1_score, confusion_matrix, classification_report
from sklearn.model_selection import learning_curve
```

```python
X_train.shape
```

```python
def evaluation(model):
    model.fit(X_train, y_train)
    y_pred = model.predict(X_test)
    
    N, train_score, val_score = learning_curve(model, X_train, y_train, cv=4,
                                              train_sizes=np.linspace(0.1, 1.0, 5))
    
    plt.figure(figsize=(12, 8))
    plt.plot(N, train_score.mean(axis=1), label="Train Score")
    plt.plot(N, val_score.mean(axis=1), label="Validation Score")
    plt.legend()
    plt.show()
```

```python
evaluation(SGD)
```

```python
from sklearn.preprocessing import PolynomialFeatures, StandardScaler
from sklearn.feature_selection import SelectKBest, f_classif
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import AdaBoostRegressor, RandomForestRegressor
from sklearn.neighbors import KNeighborsRegressor
```

```python
preprocessor = make_pipeline(PolynomialFeatures(2), SelectKBest(f_classif, k=10))
```

```python
DecisionTree = make_pipeline(preprocessor, StandardScaler(), DecisionTreeRegressor())
AdaBoost = make_pipeline(preprocessor, StandardScaler(), AdaBoostRegressor())
RandomForest = make_pipeline(preprocessor, StandardScaler(), RandomForestRegressor())
KNN = make_pipeline(preprocessor, StandardScaler(), KNeighborsRegressor())
```

```python
dict_of_models = {'DecisionTree': DecisionTree,
                 'AdaBoost': AdaBoost,
                 'RandomForest': RandomForest,
                 'KNN': KNN}
```

```python
for name, model in dict_of_models.items():
    print(name)
    evaluation(model)
```

# Optimisation

```python
from sklearn.model_selection import GridSearchCV, RandomizedSearchCV
```

```python
SGD.get_params(deep=True)
```

```python
hyper_params = {'sgdregressor__alpha': [0.0001, 0.001, 0.002, 0.01, 0.02],
               'sgdregressor__epsilon': [0.2, 0.5, 0.6, 0.7],
               'sgdregressor__l1_ratio': [0.1, 0.15, 0.2, 0.25],
               'sgdregressor__n_iter_no_change': [3, 4, 5, 6, 7],
               'sgdregressor__tol': [0.001, 0.1, 0.02, 0.15],
               'sgdregressor__validation_fraction': [0.1, 0.2, 0.3, 0.001]}
```

```python
grid = RandomizedSearchCV(SGD, hyper_params, cv=5, n_iter=50)
```

```python
grid.fit(X_train, y_train)
print(grid.best_params_)
y_pred = grid.predict(X_test)
```

```python
grid.score(X_test, y_test)
```

```python
evaluation(grid.best_estimator_)
```

```python
model = grid.best_estimator_
model.score(X_test, y_test)
```
